```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('covid_19_clean_complete.csv')
```

```python
data.info()
```

```python
data
```

```python
data["Province/State"].nunique()
```

```python
plt.xticks(rotation=6, horizontalalignment='left')
sns.lineplot(x='Confirmed',y='Deaths',data=data)
```

```python
covid = data.pivot_table(index='Country/Region',columns='Date',values='Confirmed')
```

```python
covid
```

```python
sns.heatmap(covid, annot=True)
```

```python
sns.heatmap(covid,cmap="coolwarm")
```

```python
sns.rugplot(data['Confirmed']) 
```

```python
sns.rugplot(data['Deaths']) 
```

```python
sns.rugplot(data['Recovered'])
```

```python
#matplotlib
x = data["Confirmed"] 
y = data["Deaths"] 
plt.scatter(x, y, color= "r", marker= ".")  
plt.xlabel('Cases') 
plt.ylabel('Deaths') 
plt.tight_layout()
plt.show() 
```

```python
sns.lmplot(x="Confirmed", y="Deaths", data=data)
```

```python
x = data["Confirmed"] 
y = data["Deaths"] 
plt.xlabel('Confirmed') 
plt.ylabel('Deaths') 
plt.stem(x, y, use_line_collection=True)
plt.show()
```

```python
#Plotting Power Spectral Density (PSD) in Matplotlib.
s = data["Confirmed"]
p = data["Deaths"]
d = data["Recovered"]
plt.psd(s)

plt.psd(p)

plt.psd(d)

plt.show()
```
